```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1, 9, 9).reshape(-1, 1)
y = np.array([52, 191, 350, 571, 912, 1207, 1682, 2135, 2684]).reshape(-1, 1)
```

```python
plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(x, y)
```

```python
x1 = np.array([10]).reshape(-1, 1)
lin_reg.predict(x1)
```
